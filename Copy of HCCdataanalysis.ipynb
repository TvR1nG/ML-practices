```
import numpy as np

array=np.array([[1,2,3],[2,3,4]],dtype=np.int32)
a=np.zeros((3,4))
ab=np.ones((3,4))


print(array)
print(array.dtype)
print('dimension',array.ndim)
print('shape',array.shape)
print('size',array.size)
print(a)
print(ab)
```

```
aa=np.arange(12)
ac=np.arange(12).reshape((3,4))
ad=np.linspace(1,10,4).reshape((2,2))

print(ac)
print(ad)
print(ad<3)
```

```
aaa=np.array([[1,1],[0,1]])
bbb=np.arange(4).reshape((2,2))
#矩阵乘法 
ccc=np.dot(aaa,bbb)
c_dot=aaa.dot(bbb)

print(ccc)
print(c_dot)
```

```
aab=np.random.random((2,4))

print(aab)
print(np.sum(aab,axis=1))
#axis=1，列数求和
#axis=0，行数求和
print(np.min(aab,axis=0))
print(np.max(aab))
```
